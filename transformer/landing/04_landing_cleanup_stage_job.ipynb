# 04_landing_cleanup_stage_job
---
Este notebook remove os arquivos **csv** e **parquet** da camada **Stage** após a conclusão do carregamento na camada **Bronze**.


```python
# Parameters

stage_path = "/opt/airflow/data-layer/stage"
```

```python
import os
from pyspark.sql import SparkSession
from transformer.utils.spark_helpers import get_spark_session
from transformer.utils.file_io import check_files_in_folder, delete_files
from transformer.utils.logger import get_logger

log = get_logger("landing.cleanup_stage")

spark = get_spark_session("CleanupStage")
log.info("[Landing][CleanupStage] SparkSession iniciada.")
```

```python
try:
    log.info("[Landing][CleanupStage] Iniciando job de limpeza da stage.")

    removable_files = check_files_in_folder(stage_path, "*.*")
    if not removable_files:
        log.warning("[Landing][CleanupStage] Nenhum arquivo encontrado para remoção.")
    else:
        delete_files(spark, removable_files)
        log.info(f"[Landing][CleanupStage] {len(removable_files)} arquivo(s) removido(s) da stage.")

    log.info("[Landing][CleanupStage] Job concluído com sucesso.")

except Exception as e:
    log.exception(f"[Landing][CleanupStage][ERROR] Falha durante execução: {e}.")
    raise
```

```python
# Encerrando a sessão Spark
spark.stop()
log.info("[Landing][CleanupStage] Sessão Spark finalizada.")
```
